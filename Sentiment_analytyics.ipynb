Before you can make any API calls, you need to initialize the Generative Model. I'll use `gemini-1.5-pro` for this task.

```python
# Using the verified 'latest' alias to resolve 404 errors
gemini_model = genai.GenerativeModel('gemini-flash-latest')
print(f"Active model initialized: {gemini_model.model_name}")
```

```python
df.loc[50]
```

Now, let's pick a sample tweet from the loaded dataset and use the Gemini model to classify its sentiment based on the rules you provided.

```python
try:
    sample_tweet = df.loc[50]
    entity = sample_tweet['Entity']
    tweet_text = sample_tweet['Tweet Content']

    prompt = f"""You are a sentiment classification assistant.
Task: Classify the sentiment of the tweet toward {entity} as Positive, Negative, or Neutral.

Entity: {entity}
Tweet: \"{tweet_text}\"

Sentiment:"""

    response = gemini_model.generate_content(prompt)
    print('Model Response:', response.text)
    if response.text:
        sentiment = response.text.strip()
        print(f"--- Tweet Content ---\n{tweet_text}")
        print(f"\nClassified Sentiment for {entity}: {sentiment}")
    else:
        print("Model returned an empty response.")

except Exception as e:
    print(f"An error occurred during classification: {e}")
```
